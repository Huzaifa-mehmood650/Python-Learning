```python
import pandas as pd
data = {
    "Match_ID": [1001, 1002, 1003, 1004, 1005],
    "Date": ["2024-02-17", "2024-02-18", "2024-02-19", "2024-02-20", "2024-02-21"],
    "Venue": ["Lahore", "Karachi", "Rawalpindi", "Multan", "Lahore"],
    "Team 1": ["Lahore Qalandars", "Karachi Kings", "Islamabad United", "Peshawar Zalmi", "Multan Sultans"],
    "Team 2": ["Multan Sultans", "Peshawar Zalmi", "Karachi Kings", "Islamabad United", "Lahore Qalandars"],
    "Winner": ["Multan Sultans", "Peshawar Zalmi", "Islamabad United", "Peshawar Zalmi", "Multan Sultans"],
    "Toss Winner": ["Multan Sultans", "Karachi Kings", "Islamabad United", "Peshawar Zalmi", "Multan Sultans"],
    "Toss Decision": ["Bat", "Field", "Bat", "Bat", "Field"],
    "Umpire 1": ["Aleem Dar", "Ahsan Raza", "Rashid Riaz", "Aleem Dar", "Asif Yaqoob"],
    "Umpire 2": ["Asif Yaqoob", "Aleem Dar", "Ahsan Raza", "Rashid Riaz", "Ahsan Raza"]
}
data = pd.DataFrame(data)
print(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
win_counts = data['Winner'].value_counts()
win_counts.plot(kind='bar', figsize=(10, 6), color='blue')
plt.title("Number of Matches Won by Each Team")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
win_counts = data['Winner'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(win_counts, labels=win_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("PSL Match Wins by Team")
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Team': [
        'Islamabad United', 'Karachi Kings', 'Lahore Qalandars',
        'Peshawar Zalmi', 'Quetta Gladiators', 'Multan Sultans'
    ],
    'Matches Played': [90, 85, 87, 92, 88, 65],
    'Wins': [50, 35, 40, 48, 41, 38]
}
df = pd.DataFrame(data)
df['Win %'] = (df['Wins'] / df['Matches Played']) * 100
plt.figure(figsize=(8, 8))
plt.pie(
    df['Win %'],
    labels=df['Team'],
    autopct='%1.1f%%',
    startangle=140,
    shadow=True
)
plt.title('PSL Teams Winning Percentage')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Runs': [180, 145, 200, 170, 160],
    'Wickets': [6, 8, 5, 7, 9],
    'Catches': [2, 3, 1, 4, 3],
    'Sixes': [12, 8, 15, 10, 11]
}
df = pd.DataFrame(data)
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
